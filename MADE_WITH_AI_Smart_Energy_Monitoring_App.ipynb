```python
import gradio as gr
import pandas as pd
import numpy as np

from sklearn.ensemble import RandomForestRegressor

import plotly.express as px
```

```python
np.random.seed(42)

rows = 1000

voltage = np.random.uniform(220,240,rows)

current = np.random.uniform(1,20,rows)

temperature = np.random.uniform(10,40,rows)

hour = np.random.randint(0,24,rows)

power = (
    voltage
    * current
    * (1 + temperature/100)
)

df = pd.DataFrame({
    "voltage": voltage,
    "current": current,
    "temperature": temperature,
    "hour": hour,
    "power": power
})

df.head()
```

```python
X = df[
    ["voltage","current","temperature","hour"]
]
y = df["power"]

model = RandomForestRegressor(
    n_estimators=100,
    random_state=42
)

model.fit(X,y)
```

```python
history = []

def predict_energy(
    voltage,
    current,
    temperature,
    hour
):
    prediction = model.predict([[
        voltage,
        current,
        temperature,
        hour
    ]])[0]

    history.append({
        "Voltage": voltage,
        "Current": current,
        "Temperature": temperature,
        "Hour": hour,
        "Predicted Power": prediction
    })

    return round(prediction,2)
```

```python
def analytics_dashboard():

  if len(history) ==0:
    return None

  df_history = pd.DataFrame(history)

  fig = px.line(
      df_history,
      y="Predicted Power",
      title="Prediction History"
  )

  return fig
```

```python
with gr.Blocks() as app:

  gr.Markdown(
      "# Smart Energy Prediction Platform"
  )

  voltage = gr.Slider(
      220,
      240,
      label="Voltage"
  )

  current = gr.Slider(
      1,
      20,
      label="Current"
  )

  temperature = gr.Slider(
      10,
      40,
      label = "Temperature"
  )

  hour = gr.Slider(
      0,
      23,
      label = "Hour"
  )

  prediction = gr.Textbox(
      label="Predicted Power"
  )

  predict_btn = gr.Button(
      "Predict"
  )

  chart = gr.Plot()

  predict_btn.click(
      predict_energy,
      inputs=[
          voltage,
          current,
          temperature,
          hour
      ],
      outputs=prediction
  )

  predict_btn.click(
      analytics_dashboard,
      outputs=[
          chart
      ]
  )
```

```python
app.launch(
    share=True
)
```
